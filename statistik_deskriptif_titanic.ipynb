# 3.0 | membaca file dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


dataset = pd.read_csv('titanic.csv')
```

# 3.1 | Read dataset titanic yang disematkan pada assignment ini lalu tampilkan 5 data teratas dan 5 data terbawah

```python
display(dataset)
```

# 3.2 | Gunakan fungsi deskripsi untuk melakukan statistik deskripsi pada dataset

```python
dataset.describe()
```

# 3.3 | Visualisasikan distribusi statistik data "Age" untuk melihat min, max, q1, q2, q3, outlier, dan rata-rata

```python
ds_age = dataset['Age'].describe()
print(ds_age.agg(['min', 'max', '25%', '50%', '75%', 'mean']))

# Plotting Nilai
plt.figure(figsize=(12, 6))
sns.boxplot(x=dataset['Age'])
plt.title('Distribusi Statistik Data "Age"')
plt.show()
```

# 3.4 | Visualisasikan frekuensi "Passenger Fare" untuk melihat distribusi penumpang

```python
ds_pasfare = dataset['Passenger Fare']

#Frekuensi muncul sesuai value harga
freq = ds_pasfare.value_counts()
print(freq)

# Plotting nilai
plt.figure(figsize=(12, 6))
plt.plot(freq, marker="o")
plt.title('Visual Frekuensi "Passenger Fare"')
plt.xlabel('Passenger Fare')
plt.ylabel('Frekuensi')
plt.show()
```

# 3.5. | Visualisasikan perbandingan antara "Sex" dan "Survived" untuk melihat mana yang paling banyak selamat dan mana yang paling banyak tidak selamat.

```python
# Membuat visualisasi
plt.figure(figsize=(12, 6))
sns.countplot(x='Sex', hue='Survived', data=dataset, palette='cool')

# Memberi label pada grafik
plt.title('Hubungan Antara Jenis Kelamin dan Kelangsungan Hidup di Titanic')
plt.xlabel('Sex')
plt.ylabel('Survived')
plt.legend(loc='upper center', labels=['Tidak Selamat', 'Selamat'])
plt.show()
```

# 3.6. Visualisasikan rombongan mana yang "Semuanya selamat", "Sebagian selamat", dan "Semuanya tidak selamat" yang dikelompokkan berdasarkan "Ticket Number"

```python
ds_ticket = dataset['Ticket Number']

# Mengelompokan orang yang selamat berdasarkan ticket number dan di jumlah sesuai dengan ticket numbernya 
ds_korelasi = dataset.groupby('Ticket Number')['Survived'].sum().reset_index()
df = pd.DataFrame(ds_korelasi)

# Melakukan kategorisasi "Semua Selamat", "Sebagian Selamat", dan "Semua Tidak Selamat"
kondisi = [df['Survived'].str.count('No')==0, df['Survived'].str.contains('NoYes') | df['Survived'].str.contains('YesNo'), df['Survived'].str.count('Yes')==0]
kategori = ['Selamat', 'Sebagian', 'Tidak Selamat']

# Membuat kolom baru bernama rombongan
df['Rombongan'] = np.where(df['Survived'].str.count('No')==0, 'Selamat',
                           (np.where(df['Survived'].str.contains('NoYes') | df['Survived'].str.contains('YesNo'), 'Sebagian Selamat',
                           'Tidak Selamat')))
# Output Data
display(df)

# Membuat visualisasi
plt.figure(figsize=(12, 6))
sns.countplot(x='Rombongan', hue='Ticket Number', data=df, palette='cool')

# Memberi label pada grafik
plt.title('Rombongan Selamat, Sebagian Selamat dan Tidak Selamat berdasarkan Ticket Number')

# Legend dihilangkan karena ticket number panjang 
plt.legend('')
# plt.legend(loc='lower right')
plt.show()
```
